```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
print("\nMissing values after handling:\n", df.isnull().sum())
```

```python
df['Survived'] = df['Survived'].astype('category')
```

```python
df['Pclass'] = df['Pclass'].astype('category')
```

```python
df['Sex'] = df['Sex'].astype('category')
```

```python
df['Embarked'] = df['Embarked'].astype('category')
```

```python
print(df.dtypes)
print("\nPreview of cleaned dataset:\n", df.head())
```

```python
print("Missing values after handling:\n", df.isnull().sum())
```

```python
print("Dataset Summary:\n", df.describe(include="all"))
```

```python
df['Survived'] = df['Survived'].astype(int)
```

```python
print("\n Survival Rate by Gender:\n", 
      df.groupby("Sex")['Survived'].mean())
```

```python
print("\n Survival Rate by Class:\n", 
      df.groupby("Pclass")['Survived'].mean())
```

```python
print("\n Survival Rate by Embarked Port:\n", 
      df.groupby("Embarked")['Survived'].mean())
```

```python
print("\n Average Age by Survival:\n", 
      df.groupby("Survived")['Age'].mean())
```

```python
print("\n Average Fare by Survival:\n", 
      df.groupby("Survived")['Fare'].mean())
```

# 0 INDICATES NO AND 1 INDICATES YES

```python
sns.countplot(data=df, x="Survived", palette="Set1")
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(data=df, x="Sex", hue="Survived", palette="Set1")
plt.title("Survival by Gender")
plt.show()
```

```python
sns.scatterplot(data=df, x="Age", y="Fare", hue="Survived", style="Sex", palette="Set1")
plt.title("Fare vs Age colored by Survival")
plt.show()
```

```python
sns.countplot(data=df, x="Pclass", hue="Survived", palette="Set2")
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.boxplot(data=df, x="Pclass", y="Fare", hue="Survived", palette="coolwarm")
plt.title("Fare distribution across Classes and Survival")
plt.show()
```

```python
sns.histplot(data=df, x="Age", hue="Survived", bins=30, kde=True, palette="coolwarm")
plt.title("Age Distribution by Survival")
plt.show()
```

```python
g = sns.FacetGrid(df, col="Survived", row="Sex", margin_titles=True)
g.map_dataframe(sns.histplot, x="Age", bins=20)
g.set_axis_labels("Age", "Count")
plt.show()
```

```python
sns.boxplot(data=df, x="Survived", y="Fare", palette="coolwarm")
plt.title("Fare Distribution by Survival")
plt.show()
```

```python
sns.countplot(data=df, x="Embarked", hue="Survived", palette="pastel")
plt.title("Survival by Embarkation Port")
plt.show()
```

```python
sns.pairplot(df[["Survived", "Age", "Fare", "Pclass"]], hue="Survived", palette="Set1")
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()
```

```python
pivot = df.pivot_table(values="Survived", index="Sex", columns="Pclass", aggfunc="mean")
sns.heatmap(pivot, annot=True, cmap="Blues", fmt=".2f")
plt.title("Survival Rates by Gender and Class")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
